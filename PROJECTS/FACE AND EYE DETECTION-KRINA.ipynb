```python
import numpy as np
import cv2
 
#cv2.cascadeclassifier helps to load the xml features
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_eye.xml')
#face_classifier=helps to dtermine where is the face in thhsi particular image
#eye_classifier=" eye in image

img = cv2.imread('C:\\Users\\krina\\kavita.jpg') #reading some image 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #convert the image into gray color from bgr(bluegreenred)
#gray channel easy to process  black and white , processing is quicker

faces = face_classifier.detectMultiScale(gray, 1.3, 5)
#initially we dtect face
#detectmultiscale helps to findout feature and location in the image
#gives 4 values x and y , width and height 

# When no faces detected, face_classifier returns and empty tuple
if faces is ():
    print("No Face Found")

for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(127,0,255),2) #read coordinates ad draw rectangle #and .. are the colors
    cv2.imshow('img',img)
    cv2.waitKey(0)
    roi_gray = gray[y:y+h, x:x+w] #then we crop the particular face
    roi_color = img[y:y+h, x:x+w] #image details 
    eyes = eye_classifier.detectMultiScale(roi_gray) #detect multiscale feature from roi_gray and detects eye from the particular face
    for (ex,ey,ew,eh) in eyes: #again 4 coordinates x,y,weight,height
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,255,0),2) #GAIN DRAW RECTANGLE
        cv2.imshow('img',img)
        cv2.waitKey(0)
    
cv2.destroyAllWindows()

#initially coordinates x,y is the point then adding heught and width
```

```python
# Face Recognition AND EYE FROM VIDEOES( SHOULD BE ABLE TO DETERMINE EYE AND FACE FETAURES)
#OPEN WEBCAM USIN OPENCV  (LAPTOPS HAVE INTEGRATED WEBCAM)

# Importing the libraries
import cv2

# Loading the cascades #FRONT FACE AND EYE DETERMINE FROM VIDEOS
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_eye.xml')

# Defining a function that will do the detections
def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray, 1.3, 5) #HELPS TO SEE FACE FEATURES FROM GRAY IMAGE
    for (x, y, w, h) in faces: #4 COORDINATES
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2) #DRAW RECTANGEL ON FRAME
        roi_gray = gray[y:y+h, x:x+w] #CROP FACE
        roi_color = frame[y:y+h, x:x+w] #CROP FRAME
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 3) #APPLY EYE CASCADE
        for (ex, ey, ew, eh) in eyes: #DRAW RECTANGLE ON EYE
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)
    return frame #RETURN FRAME

# Doing some Face Recognition with the webcam
video_capture = cv2.VideoCapture(0) #IF WE GIVE VALUE 0 THEN DEFAULT WEBCAM WILL GET OPEN(ACTIVATED)
#WE CAN CAPTURE FRAMES AND THAT FRAME GETS CAPTURED AND OBJECT.READ WILL READ FRAMES FROM THAT VIDEO
#CONVERTING FRAMES TO GRAY ,THEN CALLING DTECT FUNCTION TAKES GRAY FRAMES 

while True:
    _, frame = video_capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(gray, frame)
    cv2.imshow('Video', canvas) #CALLING THE VIDEO DSIPLAYED
    if cv2.waitKey(1) & 0xFF == ord('q'): #ENTER OR Q KEY IT SHOULD NOT BREAK
        break
video_capture.release()
cv2.destroyAllWindows()
```
